# Query 2.9
Trade Calls - Using Simple Moving Averages. 
Plot the 21 day and 34 day Moving average with the average price and decide a Call ! 
Call should be buy whenever the smaller moving average (21) crosses over longer moving average (34) AND the call should be sell whenever smaller moving average crosses under longer moving average. 
One of the most widely used technical indicators.

```python
day21_roll=data['Average Price'].rolling(21).mean()
day34_roll=data['Average Price'].rolling(34).mean()
```

```python
fig,ax=plt.subplots(figsize=(12,8))
day21_roll.plot(ax=ax,label='21day_rolling',color='black')
day34_roll.plot(ax=ax,label='34day_rolling')
signals = pd.DataFrame(index=data.index)
signals['signal'] = 0.0
signals['21day_roll'] = day21_roll
signals['34_dayroll'] = day34_roll
signals['signal'][0:] = np.where(signals['21day_roll'][0:] 
                                            > signals['34_dayroll'][0:], 1.0, 0.0)   

signals['positions'] = signals['signal'].diff()
data['Close Price'].plot(ax=ax,label='daily_close')


ax.plot(signals.loc[signals.positions == 1.0].index, 
         signals['21day_roll'][signals.positions == 1.0],
         '^', markersize=10,label='buy', color='g')
         
ax.plot(signals.loc[signals.positions == -1.0].index, 
         signals['21day_roll'][signals.positions == -1.0],
         'v', markersize=10,label='sell', color='k')
ax.set(xlabel="Date",ylabel="Price in Rupees")
ax.legend()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.setp(ax.get_xticklabels(),rotation=45);
```

# Query 2.10
Trade Calls - Using Bollinger Bands 
Plot the bollinger bands for this stock - the duration of 14 days and 2 standard deviations away from the average 
The bollinger bands comprise the following data points- 
The 14 day rolling mean of the closing price (we call it the average) 
Upper band which is the rolling mean + 2 standard deviations away from the average. 
Lower band which is the rolling mean - 2 standard deviations away from the average. 
Average Daily stock price.
Bollinger bands are extremely reliable , with a 95% accuracy at 2 standard deviations , and especially useful in sideways moving market. 
Observe the bands yourself , and analyse the accuracy of all the trade signals provided by the bollinger bands. 
Save to a new csv file. 

```python
middle_band = data['Close Price'].rolling(window=14).mean()
std = data['Close Price'].rolling(window=14).std()
upper_band = middle_band + 2* std
lower_band = middle_band - 2* std
data['upper_bb']=upper_band.dropna()
data['lower_bb']=lower_band.dropna()
data['middle_bb']=middle_band.dropna()
```

```python
fig=plt.figure(figsize=(15,8))
ax=upper_band.plot(label='upper')
lower_band.plot(ax=ax,label='lower')
middle_band.plot(ax=ax,label='sma')
data['Average Price'].plot(ax=ax,label='actual close',linewidth=0.5)
ax.set(xlabel='Date')
ax.set_title(label='Trade Calls - Using Bollinger Bands ')
ax.legend()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.setp(ax.get_xticklabels(),rotation=45);
```

Bollinger bands are based on the concept of volatility.
Drop in price is observed whenever the space between the moving average plots i.e., increases and vice versa.
Increase in volatility means an unreliable stock and in turn leads to selling/drop call which leads to decrease in value of the stock.
